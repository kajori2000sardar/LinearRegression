```python
import numpy as np
from sklearn.linear_model import LinearRegression
import random
from matplotlib import pyplot as plt
```

```python
c=3
m=2
x=np.arange(1,101,1)
y=m*x+c
plt.scatter(x,y)
```

```python
def noise():
    return (random.random()-0.5)*33
```

```python
x_noise=[]
for i in range(len(x)):
    x_noise+=[[x[i]+noise()]]
y_noise=[]
for i in range(len(x)):
    y_noise+=[y[i]+noise()]
plt.scatter(x_noise,y_noise)
```

```python
reg = LinearRegression()
reg.fit(x_noise,y_noise)
```

```python
test=[[-10],[110]]
plt.plot(test,reg.predict(test))
plt.scatter(x_noise,y_noise)
```
